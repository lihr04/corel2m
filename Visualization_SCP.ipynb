```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
num_task=20
epochs=20

def get_mean_acc(acc,epochs):
    num_task=len(acc)
    temp=np.zeros((num_task,len(acc[0])))
    for t,v in acc.items():
        temp[t,range(t * epochs, num_task * epochs)]=v
        if t<num_task-1:
            temp[t+1,:]=temp[:t+1,:].mean(0)
    return temp.mean(0)

def get_mean_acc_on_iterations(acc_list):
    acc = {}
    for i in acc_list[0].keys():
        acc[i] = np.zeros(len(acc_list[0][i]))
    for t in range(len(acc_list)):
        for i in acc_list[0].keys():
            acc[i] += np.array(acc_list[t][i]) / len(acc_list)
    return acc
```

```python
acc_scp_list = [[] for i in range(5)]
seed_list = [1234, 2345, 3456, 4567, 5678]
for i in range(5):
    seed = seed_list[i]
    path_scp = '.\\perm_mnist_5Run\\experiment_SCP_id_%d.pkl'%(seed)
    with open(path_scp, 'rb') as f:
        _, acc_scp, hyperparameter_scp = pickle.load(f)
        for j in range(5):
            acc_scp_list[j].append(acc_scp[j])
```

```python
for j in range(5):
    acc_scp_list[j] = get_mean_acc_on_iterations(acc_scp_list[j])
```

```python
plt.figure(figsize=(10,5))
for t in range(num_task):
    if t%2:
        c='b'
    else:
        c='r'
    plt.axvspan(t*epochs, (t+1)*epochs, facecolor=c, alpha=0.1)
for i in range(5):
    plt.plot(get_mean_acc(acc_scp_list[i],epochs),linewidth=3)
plt.xticks(fontsize=18),plt.yticks(fontsize=18)
plt.legend(['1e4', '1e5', '1e6', '1e7', '1e8'],fontsize=18)
plt.title('Grid search for importance in SCP',fontsize=18)
plt.show()
```
